```python
import zipfile
from fastkml import kml
from pathlib import Path

def read_kmz(loc:Path|str):
    kmz = zipfile.ZipFile(loc, 'r')
    kml_content = kmz.open('doc.kml', 'r').read()
    print(kml_content[:10])
    k = kml.KML()
    k.from_string(kml_content.encode())
    return k
```

```python
import re
def find_source_urls(text):
    # Define a regex pattern to match the section between "Source(s):" and "Geolocation(s):"
    pattern = r"Source\(s\):\s*(.*?)\s*Geolocation\(s\):"
    
    # Search for the pattern in the text
    match = re.search(pattern, text, re.DOTALL)
    
    if match:
        # Extract the matched group, which should contain the URLs
        matched_text = match.group(1)
        # Find all URLs within the extracted text
        urls = re.findall(r"https?://[^\s]+", matched_text)
        return urls
    else:
        return []
```

```python
import json
import tqdm
for ft in tqdm.tqdm(all_feats[:1000]):
    urls = find_source_urls(ft.description)
    for url in urls:
        if "twitter.com" in url:
            try:
                twt_id = url.split("/")[-1]
                if os.path.exists(f"./tweets/{twt_id}.json"):
                    continue
                tweet = client.get_tweet_by_id(twt_id)
                dat = {
                    "text": tweet.text,
                    "urls": tweet.urls,
                    "media": tweet.media
                }
                json.dump(dat, open(f"./tweets/{twt_id}.json", "w"))
            except Exception as e:
                print(e)
```

```python
s = 0
lnks = []
fout = open("lnks.txt", 'w')
for f in all_feats:
    fout.write("-"*30 + "\n")
    fout.write(f.description + "\n")
    fout.write("DATA:" + "\n")
    for u in find_source_urls(f.description): 
        fout.write(u + "\n")
    fout.write("\n")
```
